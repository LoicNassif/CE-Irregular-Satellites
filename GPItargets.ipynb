```python
import numpy as np
def splittype(row):
    try:
        row['Class'] = row["S_Type"][0]
        row['subtype'] = row["S_Type"][1]
    except:
        row['Class'] = np.nan
        row['subtype'] = np.nan
    return row
```

```python
res = mag.apply(splittype, axis=1)
res.tail()
```

```python
Mdwarfs = res[res['Class'] == 'M']
Mdwarfs.shape
```

```python
def params(row):
    if row['subtype'] == '0':
        row['Mass'] = 0.6
        row['Lum'] = 0.072
    if row['subtype'] == '1':
        row['Mass'] = 0.49
        row['Lum'] = 0.035
    if row['subtype'] == '2':
        row['Mass'] = 0.44
        row['Lum'] = 0.023
    if row['subtype'] == '3':
        row['Mass'] = 0.36
        row['Lum'] = 0.015
    if row['subtype'] == '4':
        row['Mass'] = 0.2
        row['Lum'] = 0.0055  
        
    if np.isnan(row['FLUX_I']):
        row['contrast_thresh'] = 3e-6 # assume worst case
    else:
        imag = row['FLUX_I']
        if imag > 6 and imag < 7:
            row['contrast_thresh'] = 3e-7
        if imag > 7 and imag < 8:
            row['contrast_thresh'] = 7e-7
        if imag > 8 and imag < 9:
            row['contrast_thresh'] = 2e-6
        if imag > 9:
            row['contrast_thresh'] = 3e-6
            
    return row
```

```python
Mdwarfs = Mdwarfs.apply(params, axis=1)
```
