```python
import numpy as np
def splittype(row):
    try:
        row['Class'] = row["S_Type"][0]
        row['subtype'] = row["S_Type"][1]
    except:
        row['Class'] = np.nan
        row['subtype'] = np.nan
    return row

mag = mag.apply(splittype, axis=1)
mag
```

```python
obs = mag[mag['FLUX_I'] < 9.5]
obs
```

```python
def params(row):
    if row['Class'] == 'M':
        if row['subtype'] == '0':
            row['Mass'] = 0.6
            row['Lum'] = 0.072
            if row['Fe_H_Teff'] != 0:
                row['Teff'] = row['Fe_H_Teff']
            else:
                row['Teff'] = 3800
        if row['subtype'] == '1':
            row['Mass'] = 0.49
            row['Lum'] = 0.035
            if row['Fe_H_Teff'] != 0:
                row['Teff'] = row['Fe_H_Teff']
            else:
                row['Teff'] = 3600
        if row['subtype'] == '2':
            row['Mass'] = 0.44
            row['Lum'] = 0.023
            if row['Fe_H_Teff'] != 0:
                row['Teff'] = row['Fe_H_Teff']
            else:
                row['Teff'] = 3400
        if row['subtype'] == '3':
            row['Mass'] = 0.36
            row['Lum'] = 0.015
            if row['Fe_H_Teff'] != 0:
                row['Teff'] = row['Fe_H_Teff']
            else:
                row['Teff'] = 3250
        if row['subtype'] == '4':
            row['Mass'] = 0.2
            row['Lum'] = 0.0055  
            if row['Fe_H_Teff'] != 0:
                row['Teff'] = row['Fe_H_Teff']
            else:
                row['Teff'] = 3100
    if row['Class'] == 'K':
        if row['subtype'] == '3': # HR 8832
            row['Mass'] = 0.81
            row['Lum'] = 0.26
            row['Teff'] = row['Fe_H_Teff']
        if row['subtype'] == '5': # K2-216
            row['Mass'] = 0.7
            row['Lum'] = 0.19
            row['Teff'] = row['Fe_H_Teff']
        if row['subtype'] == '6': # GJ 9827
            row['Mass'] = 0.667
            row['Lum'] = 0.12
            row['Teff'] = row['Fe_H_Teff']
        
    if np.isnan(row['FLUX_I']):
        row['contrast_thresh'] = 3e-6 # assume worst case
    else:
        imag = row['FLUX_I']
        if imag > 6 and imag < 7:
            row['contrast_thresh'] = 3e-7
        if imag > 7 and imag < 8:
            row['contrast_thresh'] = 7e-7
        if imag > 8 and imag < 9:
            row['contrast_thresh'] = 2e-6
        if imag > 9:
            row['contrast_thresh'] = 3e-6
            
    return row
```

```python
twa = obs.apply(params, axis=1)
twa[['TWA', 'ID', 'S_Type', 'Class', 'subtype', 'FLUX_I', 'FLUX_H', 'Distance_distance', 'Mass', 'Lum', 'Teff', 'contrast_thresh']]
```
