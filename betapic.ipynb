```python
import pandas as pd
from astroquery.simbad import Simbad
```

```python
customSimbad = Simbad()
customSimbad.remove_votable_fields('coordinates')
customSimbad.get_votable_fields()
```

```python
customSimbad.add_votable_fields('ra(2;A;ICRS;J2017.5;2000)', 'dec(2;D;ICRS;2017.5;2000)', 'sp', 'distance', 'diameter', 'flux(I)', 'flux(H)', 'fe_h')
```

```python
# From Zuckerman 2001
names = ['GJ 3305', 'HD155555C', 'HIP 102409', 'HIP 29964', 'BD -176128', 'HIP 23309', 'AU Mic', 'CD -641208', 'HIP 102141']
```

```python
res = customSimbad.query_object(names[0]).to_pandas()
res
```

```python
for name in names[1:]:
    try:
        new = customSimbad.query_object(name).to_pandas()
        res = res.append(new, ignore_index=True)
    except:
        print(name, 'failed')
```

```python
res
```

```python
def bytes_to_string(row):
    row['S_Type'] = row['SP_TYPE'].decode('ascii')
    row['ID'] = row['MAIN_ID'].decode('ascii')
    return row

res = res.apply(bytes_to_string, axis=1)
res
```

```python
mag = res[['ID', 'RA_2_A_ICRS_J2017_5_2000', 'DEC_2_D_ICRS_2017_5_2000', 'S_Type', 'Distance_distance', 'FLUX_I', 'FLUX_H', 'Fe_H_Teff']]
mag
```

```python
import numpy as np
def splittype(row):
    try:
        row['Class'] = row["S_Type"][0]
        row['subtype'] = row["S_Type"][1]
    except:
        row['Class'] = np.nan
        row['subtype'] = np.nan
    return row

mag = mag.apply(splittype, axis=1)
mag
```

```python
mag = mag.drop_duplicates()
mag
```

```python
Mdwarfs = pd.read_csv('GPIMdwarfs.csv', index_col=0)
Mdwarfs.merge(mag, left_on=['Distance_distance'], right_on=['Distance_distance'])
```

```python
def params(row):
    if row['Class'] == 'M':
        if row['subtype'] == '0':
            row['Mass'] = 0.6
            row['Lum'] = 0.072
            if row['Fe_H_Teff'] != 0:
                row['Teff'] = row['Fe_H_Teff']
            else:
                row['Teff'] = 3800
        if row['subtype'] == '1':
            row['Mass'] = 0.49
            row['Lum'] = 0.035
            if row['Fe_H_Teff'] != 0:
                row['Teff'] = row['Fe_H_Teff']
            else:
                row['Teff'] = 3600
        if row['subtype'] == '2':
            row['Mass'] = 0.44
            row['Lum'] = 0.023
            if row['Fe_H_Teff'] != 0:
                row['Teff'] = row['Fe_H_Teff']
            else:
                row['Teff'] = 3400
        if row['subtype'] == '3':
            row['Mass'] = 0.36
            row['Lum'] = 0.015
            if row['Fe_H_Teff'] != 0:
                row['Teff'] = row['Fe_H_Teff']
            else:
                row['Teff'] = 3250
        if row['subtype'] == '4':
            row['Mass'] = 0.2
            row['Lum'] = 0.0055  
            if row['Fe_H_Teff'] != 0:
                row['Teff'] = row['Fe_H_Teff']
            else:
                row['Teff'] = 3100
    if row['Class'] == 'K':
        if row['subtype'] == '3': # HR 8832
            row['Mass'] = 0.81
            row['Lum'] = 0.26
            row['Teff'] = row['Fe_H_Teff'] # 4834
        if row['subtype'] == '4': # Linear interp between 3 and 5
            row['Mass'] = 0.75
            row['Lum'] = 0.23
            if row['Fe_H_Teff'] != 0:
                row['Teff'] = row['Fe_H_Teff']
            else:
                row['Teff'] = 4634
        if row['subtype'] == '5': # K2-216
            row['Mass'] = 0.7
            row['Lum'] = 0.19
            row['Teff'] = 4434
        if row['subtype'] == '6': # GJ 9827
            row['Mass'] = 0.667
            row['Lum'] = 0.12
            row['Teff'] = row['Fe_H_Teff'] # 4057
        if row['subtype'] == '7': # Linear interp between 6 and M0
            row['Mass'] = 0.63
            row['Lum'] = 0.1
            row['Teff'] = 3900   
        
    if np.isnan(row['FLUX_I']):
        row['contrast_thresh'] = 3e-6 # assume worst case
    else:
        imag = row['FLUX_I']
        if imag > 6 and imag < 7:
            row['contrast_thresh'] = 3e-7
        if imag > 7 and imag < 8:
            row['contrast_thresh'] = 7e-7
        if imag > 8 and imag < 9:
            row['contrast_thresh'] = 2e-6
        if imag > 9:
            row['contrast_thresh'] = 3e-6
            
    return row
```

```python
mag = mag.apply(params, axis=1)
mag
#twa[['TWA', 'ID', 'S_Type', 'Class', 'subtype', 'FLUX_I', 'FLUX_H', 'Distance_distance', 'Mass', 'Lum', 'Teff', 'contrast_thresh']]
```
